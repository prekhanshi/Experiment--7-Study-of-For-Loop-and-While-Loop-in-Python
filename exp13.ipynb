```python
import pandas as pd
import numpy as np

data= {
    'product':['computer','printer','monitor','tablet','smartwatch','camera'],
    'price':[55000,10000,15000,5000,8000,45000],
    'units_sold':[24,50,5,34,10,60]
}
df = pd.DataFrame(data)
print(df)
```

```python
#data binning
bins = [0,5000,15000, 55000,]
labels = ['low','medium','high']
df['price_category'] = pd.cut(df['price'],bins=bins, labels=labels)
df
```

```python
#binning units sold
bins2=[0,5,30,60]
labels2=['low','medium','high']
df['sales_category'] = pd.cut(df['units_sold'],bins=bins2, labels=labels2)
df
```

```python
#data fromatting
#convert data type
print(df.dtypes)
df['units_sold'] = df['units_sold'].astype(float)
print(df.dtypes)
```

```python
#convert text to uppercase
df['product'] = df['product'].str.upper()
df
```

```python
#rounding values
df['price'] = df['price'].round(2)
df
```

```python
#additional operations
#sort data
sorted_df = df.sort_values(by='price')
print(sorted_df)
```

```python
#display unique values
print(df['price_category'].unique())
```

```python
#dataset :food delivery orders
import pandas as pd
import numpy as np
data = {
    'order_ID':['01','02','03','04','05','06','07','08'],
    'order_value':[250,450,120,600,300,150,509,220],
    'delivery_time':[30,45,35,60,20,50,75,27],
    'distance_km':[5,7,9,20,10,3,5,4]
}
df = pd.DataFrame(data)
print(df)
```

```python
#binning order value
bins = [0,100,200,400]
labels = ['low','medium','high']
df['order_category'] = pd.cut(df['order_value'],bins=bins, labels=labels)
df
```

```python
#binning delivery time
bins2=[0,30,45,70]
labels2=['fast','medium','slow']
df['delivery_category'] = pd.cut(df['delivery_time'], bins=bins2, labels=labels2)
df
```

```python
#unique categories
print(df['order_category'].unique())
```

```python
#count of categories
print(df['order_category'].value_counts())
```

```python
#sorting
print(df.sort_values(by='order_value'))
```
